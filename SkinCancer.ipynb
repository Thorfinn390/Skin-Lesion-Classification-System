```python
# @title Skin Cancer Classifier with Professional Report
%pip install -q torch torchvision pandas pillow tqdm matplotlib
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms, models
import pandas as pd
from PIL import Image
import os
import zipfile
from tqdm import tqdm
from google.colab import drive, files
from sklearn.model_selection import train_test_split
from sklearn.utils.class_weight import compute_class_weight
import matplotlib.pyplot as plt
import numpy as np
from IPython.display import display, HTML
from datetime import datetime

# @title 1. Model Definition
class SkinCancerModel(nn.Module):
    def __init__(self):
        super().__init__()
        self.base_model = models.resnet18(weights=models.ResNet18_Weights.IMAGENET1K_V1)
        for param in self.base_model.parameters():
            param.requires_grad = False
        num_features = self.base_model.fc.in_features
        self.base_model.fc = nn.Sequential(
            nn.Linear(num_features, 512),
            nn.ReLU(),
            nn.Dropout(0.5),
            nn.Linear(512, 7)
        )

    def forward(self, x):
        return self.base_model(x)

# @title 2. Data Preparation
def verify_and_extract(data_dir, image_dir):
    if os.path.exists(image_dir) and len([f for f in os.listdir(image_dir) if f.endswith('.jpg')]) > 1000:
        print("Using previously extracted images")
        return True

    zip_files = [
        f'{data_dir}/HAM10000_images_part_1.zip',
        f'{data_dir}/HAM10000_images_part_2.zip',
        f'{data_dir}/HAM10000_metadata.csv'
    ]

    missing_files = [f for f in zip_files if not os.path.exists(f)]
    if missing_files:
        print("⚠ Missing required files:")
        for f in missing_files:
            print(f"- {os.path.basename(f)}")
        print("\nPlease ensure these files are in your Google Drive's SkinCancerData folder")
        return False

    print("Extracting images...")
    os.makedirs(image_dir, exist_ok=True)
    for zip_file in zip_files[:2]:
        try:
            with zipfile.ZipFile(zip_file, 'r') as z:
                for file in tqdm(z.namelist(), desc=f"Extracting {os.path.basename(zip_file)}"):
                    if file.lower().endswith('.jpg'):
                        try:
                            z.extract(file, image_dir)
                            if '/' in file:
                                os.rename(f"{image_dir}/{file}", f"{image_dir}/{os.path.basename(file)}")
                        except:
                            continue
        except Exception as e:
            print(f"Failed to extract {zip_file}: {str(e)}")
            return False

    extracted_images = [f for f in os.listdir(image_dir) if f.lower().endswith('.jpg')]
    if len(extracted_images) < 1000:
        print(f"⚠ Only {len(extracted_images)} images extracted - expected >1000")
        return False

    print(f"✓ Successfully extracted {len(extracted_images)} images")
    return True

# @title 3. Training Function
def train_model(data_dir, image_dir, epochs=15):
    try:
        df = pd.read_csv(f'{data_dir}/HAM10000_metadata.csv')
    except Exception as e:
        print(f"Failed to load metadata: {str(e)}")
        return None

    extracted_images = {f.split('.')[0] for f in os.listdir(image_dir) if f.lower().endswith('.jpg')}
    df = df[df['image_id'].isin(extracted_images)]

    if len(df) < 1000:
        print(f"⚠ Only {len(df)} valid image-label pairs - expected >1000")
        return None

    labels = df['dx'].map({'akiec':0, 'bcc':1, 'bkl':2, 'df':3, 'mel':4, 'nv':5, 'vasc':6})
    class_weights = compute_class_weight('balanced', classes=np.unique(labels), y=labels)
    class_weights = torch.tensor(class_weights, dtype=torch.float32)
    print(f"\nClass weights for balancing: {dict(zip(['akiec','bcc','bkl','df','mel','nv','vasc'], class_weights.numpy().round(2)))}")

    class SkinDataset(Dataset):
        def __init__(self, df, is_train=True):
            self.df = df
            self.is_train = is_train
            self.label_map = {'akiec':0, 'bcc':1, 'bkl':2, 'df':3, 'mel':4, 'nv':5, 'vasc':6}
            self.train_transform = transforms.Compose([
                transforms.Resize(256),
                transforms.RandomResizedCrop(224),
                transforms.RandomHorizontalFlip(),
                transforms.RandomVerticalFlip(),
                transforms.RandomRotation(30),
                transforms.ColorJitter(brightness=0.3, contrast=0.3, saturation=0.3),
                transforms.ToTensor(),
                transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
            ])
            self.val_transform = transforms.Compose([
                transforms.Resize(256),
                transforms.CenterCrop(224),
                transforms.ToTensor(),
                transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
            ])

        def __len__(self): return len(self.df)
        def __getitem__(self, idx):
            img_path = f"{image_dir}/{self.df.iloc[idx]['image_id']}.jpg"
            img = Image.open(img_path).convert('RGB')
            label = self.label_map[self.df.iloc[idx]['dx']]
            return (self.train_transform(img) if self.is_train else self.val_transform(img), label)

    train_df, val_df = train_test_split(df, test_size=0.2, random_state=42, stratify=df['dx'])
    train_loader = DataLoader(SkinDataset(train_df, is_train=True), batch_size=64, shuffle=True)
    val_loader = DataLoader(SkinDataset(val_df, is_train=False), batch_size=64, shuffle=False)

    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    model = SkinCancerModel().to(device)
    criterion = nn.CrossEntropyLoss(weight=class_weights.to(device))
    optimizer = optim.AdamW(model.parameters(), lr=0.0001, weight_decay=1e-4)
    scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, 'max', patience=2, factor=0.5)

    best_val_acc = 0
    for epoch in range(epochs):
        model.train()
        running_loss = 0.0
        for images, labels in tqdm(train_loader, desc=f"Epoch {epoch+1}/{epochs}"):
            images, labels = images.to(device), labels.to(device)
            optimizer.zero_grad()
            outputs = model(images)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()
            running_loss += loss.item()

        model.eval()
        val_loss = 0.0
        correct = 0
        total = 0
        class_correct = [0] * 7
        class_total = [0] * 7

        with torch.no_grad():
            for images, labels in val_loader:
                images, labels = images.to(device), labels.to(device)
                outputs = model(images)
                loss = criterion(outputs, labels)
                val_loss += loss.item()
                _, predicted = torch.max(outputs.data, 1)
                total += labels.size(0)
                correct += (predicted == labels).sum().item()
                for i in range(7):
                    class_correct[i] += ((predicted == labels) & (labels == i)).sum().item()
                    class_total[i] += (labels == i).sum().item()

        val_acc = 100 * correct / total
        print(f"Train Loss: {running_loss/len(train_loader):.4f} | Val Loss: {val_loss/len(val_loader):.4f} | Val Acc: {val_acc:.1f}%")
        print("Class-wise Val Accuracy:")
        for i, (cc, ct) in enumerate(zip(class_correct, class_total)):
            if ct > 0:
                print(f"{['akiec','bcc','bkl','df','mel','nv','vasc'][i]}: {100*cc/ct:.1f}% ({ct} samples)")

        if val_acc > best_val_acc:
            best_val_acc = val_acc
            torch.save(model.state_dict(), 'skin_cancer_model.pth')
            print("↳ Saved new best model")

        scheduler.step(val_acc)

    print("\n✓ Training complete! Best model saved as 'skin_cancer_model.pth'")
    return model

# @title 4. Professional Prediction Report
def predict_and_report(image_path, model):
    """Generate professional medical report with confidence assessment"""
    class_info = {
        0: ('Actinic keratoses (akiec)',
            'Pre-cancerous lesion that may progress to squamous cell carcinoma'),
        1: ('Basal cell carcinoma (bcc)',
            'Most common skin cancer, rarely metastasizes but can be locally destructive'),
        2: ('Benign keratosis-like lesions (bkl)',
            'Benign lesions including solar lentigines and seborrheic keratoses'),
        3: ('Dermatofibroma (df)',
            'Benign fibrous skin nodule, no malignant potential'),
        4: ('Melanoma (mel)',
            'Most dangerous skin cancer, can metastasize if not treated early'),
        5: ('Melanocytic nevi (nv)',
            'Common benign mole, very low malignant potential'),
        6: ('Vascular lesions (vasc)',
            'Benign vascular proliferations including angiomas')
    }

    biopsy_recommendation = {
        0: "Strongly recommended - Due to potential progression to squamous cell carcinoma",
        1: "Recommended - Standard treatment for basal cell carcinoma",
        2: "Not typically indicated - Benign lesion",
        3: "Not indicated - Clearly benign characteristics",
        4: "Urgently recommended - Melanoma requires immediate attention",
        5: "Not indicated unless changing - Typical benign mole",
        6: "Not indicated - Clearly benign vascular lesion"
    }

    transform = transforms.Compose([
        transforms.Resize(256),
        transforms.CenterCrop(224),
        transforms.ToTensor(),
        transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
    ])

    try:
        img = Image.open(image_path).convert('RGB')
        img_tensor = transform(img).unsqueeze(0)

        device = next(model.parameters()).device
        with torch.no_grad():
            outputs = model(img_tensor.to(device))
            probabilities = torch.nn.functional.softmax(outputs, dim=1)[0] * 100
            confidence, pred = torch.max(probabilities, 0)

        diagnosis, description = class_info[pred.item()]
        biopsy = biopsy_recommendation[pred.item()]

        # Generate professional medical report
        current_date = datetime.now().strftime("%Y-%m-%d %H:%M")
        report = f"""============================================================
            SKIN LESION ANALYSIS REPORT - CONFIDENTIAL
============================================================

Patient Information:
------------------------------------------------------------
Date of Analysis: {current_date}
Image File: {os.path.basename(image_path)}
Image Dimensions: {img.size[0]} x {img.size[1]} pixels

Diagnostic Findings:
------------------------------------------------------------
Primary Diagnosis: {diagnosis}
Confidence Level: {confidence.item():.1f}% ({'High' if confidence > 75 else 'Moderate' if confidence > 50 else 'Low'} confidence)
Clinical Description: {description}

Probability Distribution:
------------------------------------------------------------
"""
        for i, prob in enumerate(probabilities):
            report += f"{class_info[i][0]:<45}: {prob:>5.1f}%\n"

        report += f"""
Clinical Recommendations:
------------------------------------------------------------
Biopsy Recommendation: {biopsy}

Next Steps:
{
    '• Urgent dermatology referral within 48 hours' if confidence >= 60 and pred.item() in [0,1,4]
    else '• Routine dermatology follow-up' if confidence >= 60
    else '• Immediate clinical evaluation recommended due to uncertain diagnosis'
}

Additional Notes:
------------------------------------------------------------
• This report is generated by an AI diagnostic assistant
• Final diagnosis requires clinical correlation
• All suspicious lesions should be evaluated by a dermatologist
• Melanoma risk factors should be assessed regardless of diagnosis

============================================================
End of Report
"""
        # Visualization
        plt.figure(figsize=(14, 6))
        plt.subplot(1, 2, 1)
        plt.imshow(img)
        plt.title(f"Clinical Image\n{diagnosis.split(' - ')[0]}")
        plt.axis('off')

        plt.subplot(1, 2, 2)
        colors = ['red' if i == pred.item() else 'blue' for i in range(7)]
        plt.barh([class_info[i][0] for i in range(7)], probabilities.cpu().numpy(), color=colors)
        plt.xlim(0, 100)
        plt.title(f"Diagnostic Confidence Distribution\n(Confidence: {confidence.item():.1f}%)")
        plt.tight_layout()
        plt.show()

        # Create downloadable report
        display(HTML(f"""
        <div style="margin:20px; font-family:Arial">
            <h3 style="color:#2c3e50">Skin Lesion Analysis Report</h3>
            <textarea id="report" style="width:100%; height:300px; font-family:monospace; padding:10px">{report}</textarea>
            <button onclick="downloadReport()" style="padding:10px; background:#2980b9; color:white; border:none; margin-top:10px">
                Download Professional Report (PDF)
            </button>
        </div>
        <script>
        function downloadReport() {{
            var text = document.getElementById("report").value;
            var blob = new Blob([text], {{type: "text/plain"}});
            var a = document.createElement("a");
            a.href = URL.createObjectURL(blob);
            a.download = "Professional_Skin_Lesion_Report_{datetime.now().strftime('%Y%m%d')}.txt";
            document.body.appendChild(a);
            a.click();
            document.body.removeChild(a);
        }}
        </script>
        """))

        return report

    except Exception as e:
        print(f"Error during analysis: {str(e)}")
        return None

# @title 5. Main Application
def main():
    print("PROFESSIONAL SKIN CANCER CLASSIFIER")
    print("-----------------------------------")

    # Mount Google Drive
    drive.mount('/content/drive', force_remount=True)
    data_dir = '/content/drive/MyDrive/SkinCancerData'
    image_dir = '/content/images'

    # Model loading
    model_path = 'skin_cancer_model.pth'
    if os.path.exists(model_path):
        print("\nLoading existing model...")
        device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
        model = SkinCancerModel().to(device)
        model.load_state_dict(torch.load(model_path, map_location=device))
        model.eval()
    else:
        print("\nNo existing model found. Training new model...")
        if not verify_and_extract(data_dir, image_dir):
            print("Data verification failed")
            return
        model = train_model(data_dir, image_dir)

    if model is None:
        print("Failed to initialize model")
        return

    # Prediction loop
    while True:
        print("\nREADY FOR ANALYSIS - Upload a skin lesion image")
        uploaded = files.upload()
        if not uploaded:
            print("No file uploaded")
            continue

        image_path = next(iter(uploaded))
        print("\nAnalyzing image and generating professional report...")
        report = predict_and_report(image_path, model)

        if not report:
            print("Analysis failed - please try another image")

        another = input("\nWould you like to analyze another image? (y/n): ").lower()
        if another != 'y':
            print("\nThank you for using the Professional Skin Cancer Classifier")
            break

if __name__ == "__main__":
    main()
```
